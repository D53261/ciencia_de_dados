# Análise exploratória de dados

Agora sera feita uma melhor analise exploratória dos dados, isto é, observar com mais aptidão o conteudo, tipos e estrutura dos dados. Ja havia sido feito algo semelhante ate agora, mas adentraremos melhor no assunto

## Qual o valor médio de aluguel por tipo de imóvel?

Iniciaremos respondendo a essa pergunta utilisando os metodos de analise exploratoria dos dados da biblioteca pandas. Primeiro usamos .head() para visualisação inicial

```python
dados.head()
```

Observando, para respondermos a pergunta usaremos mais as colunas 'Tipo' e 'Valor', se referindo aos tipos e preços de aluguel

Agora vamos calcular o valor da média dos preços de aluguel(coluna 'Valor') no geral utilisando o metodo .mean() do pandas para calculo de médias

```python
dados['Valor'].mean()
```

Vemos que recebemos a media, porem apenas recebemos a media de todos os valores presentes na coluna 'Valor', mas queremos agrupar tudo de acordo com a coluna 'Tipo'. Para isso existe o metodo .groupby() que agrupa valores de acordo com uma coluna, onde dentro dos parenteses colocamos a coluna que sera usada como base. No nosso caso utilisaremos a coluna 'Tipo' como base

```python
dados.groupby('Tipo')
```

Isso ocorreu porque o .groupby() geralmente é utilizado em conjunto com outro metodo do pandas de diferentes tipos, como o .sum() para somar ou o recentemente visto .mean() para médias, que sera utilizado agora, mas o metodo escolhido pode ser ate uma personalização do usuario. Tambem não podemos esquecer que existem alem da coluna 'Tipo' outras colunas com valores não numericos, ou seja, não calculaveis, podemos resolver isso usando dentro dos parenteses de .mean() **numeric_only=True** para garantir que somente numeros sejam inclusos no callculo

```python
dados.groupby('Tipo').mean(numeric_only=True)
```

Agora os indices do dataframe gerado são os tipos de imóveis presentes na agora coluna principal 'Tipo', apresentando alem de 'Valor' que desejavamos, outras médias de acordo com seus tipos de imovel, todos os dados obviamente em questão do dataframe original

Retornando, agora podemos selecionar apenas a coluna 'Valor' que desejamos, abrindo colchetes no mesmo codigo anterior, antes de .mean(), com 'Valor' dentro dele para nos ser devolvido uma series. Levando em conta que agora só estão presentes dados numerico sem excessão, podemos opcionalmente tirar o comando que organiza para apenas numeros sejam incluidos no calculo
 (numeric_only=True) por não ser mais necessário

```python
dados.groupby('Tipo')['Valor'].mean()
```

Conseguimos responder a pergunta, porem a visualização dos dados ficou não tão boa por se tratarem de muitos dados, podemos melhorar isso colocando na mesma linha de codigo mais um colchete em ['Valor'] para torna-lo um dataframe

Alem disso ha a possibilidade de ordena-los da menor media para a maior, com o metodo sort_values(), passando entre os parentes a coluna em que sera aplicada a ordenação, no caso 'Valor'

```python
dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

A visualização tornou-se melhor, porem é trabalhoso olhar numero por numero para analisar tudo, então um grafico deixaria tudo mais simples.

para começar é importante guardar os dados do dataframe acima em uma variavel

```python
df_preco_tipo = dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

Em seguida criaremos um grafico de barras horizontais, chamando o nome da variavel que esta com o dataframe e usando .plot() para criar o grafico, alem de que entre parenteses sera passado:



*   kind='barh': que define o tipo do grafico, no caso 'barh' significa barras horizontais
*   figsize=(14, 10): que serve para definir o tamanho do grafico
*   color ='purple': que simplesmente define a cor do grafico



```python
df_preco_tipo.plot(kind='barh', figsize=(14,10), color='purple')
```

aqui temos um grafico de barras horizontais onde o eixo x são os valores de aluguel medios para cada imovel, que sao representados pelo eixo y

## Removendo os imóveis comerciais

Como ha uma enorme variancia entre os valores, alem de possuirem tambem edificios comerciais entre os residenciais, que são o foco no momento, removeremos os edificios comerciais para fins de organização

Inicialmente usaremos o metodo .unique() para agrupar todos os elementos de um coluna determinada, aqui sendo colocado dados.Tipo.unique()

```python
dados['Tipo'].unique()
```

Nos foi retornado todos os elementos presentes na coluna 'Tipo', e se forem unificados os imoveis comerciais um por um dessa array, ficara:

```python
imoveis_comerciais = ['Conjunto Comercial/Sala',
                      'Prédio Inteiro', 'Loja/Salão',
                      'Galpão/Depósito/Armazém',
                      'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial',
                      'Box/Garagem', 'Chácara',
                      'Loteamento/Condomínio', 'Sítio',
                      'Pousada/Chalé', 'Hotel', 'Indústria']
```

Ja salvo em uma variavel, nosso objetivo agora é retirar do dataframe principal todos os itens presentes nessa lista. Utilizaremos o metodo .query() que é responsavel por selecionar linhas especificas de um dataframe de acordo com uma condição pre definida, onde é passado assim:

```python
dados.query('@imoveis_comerciais in Tipo')
```



*   dados.query: o metodo, que sempre sera chamado com a variavel do dataframe principal
*   ('@imoveis_comerciais'): sempre funcionara entre aspas e colocando um @ antes de inserir o nome da variavel desejada
*   ('in Tipo'): no mesmo parenteses sera usado um in(dentro de) e a coluna presente na variavel dados que foi chamado com o query



Nos foi retornado um dataframe oposto do que queremos, com menos linhas por apresentar apenas itens presentes na variavel imoveis_comerciais que estavam em tipo, mas para reverter isso e chegarmos onde queremos, apenas colocaremos um **not** antes de **Tipo** na expressão ('imoveis_comerciais in Tipo')

```python
dados.query('@imoveis_comerciais not in Tipo')
```

Como conseguimos o dataframe com os dados desejados e terminamos a exploração, vamos salva-lo em uma variavel 'df' e observar com .head() se tudo ocorreu bem

```python
df = dados.query('@imoveis_comerciais not in Tipo')
df.head()
```

Tambem iremos conferir com .unique() se só ha tipos de imoveis residenciais e nenhum remanescente da lista criada anteriormente de imoveis comerciais

```python
df.Tipo.unique()
```

Agora que vimos que tudo esta certo, vamos plotar novamente o grafico apenas mudando o nome das variaveis para que tudo se aplique

```python
df_preco_tipo = df.groupby('Tipo')[['Valor']].mean().sort_values('Valor')

df_preco_tipo.plot(kind='barh', figsize=(14,10), color='purple')
```

Agora, com muito menos itens na coluna 'Tipo' vemos como o grafico ficou muito mais organizado e como conseguimos responder a pergunta proferida

## Qual o percentual de cada tipo de imóvel na nossa base de dados?

Agora temos uma pergunta que se refere a quantas vezes cada elemento aparece no dataframe, para iniciarmos essa analise, iremos usar o metodo .unique() que ja nos foi apresentado para mostrar todos os elementos presentes, alem de selecionar a coluna 'Tipo' para que o codigo funcione

```python
df['Tipo'].unique()
```

agora que vimos os tipos unicos de imoveis presentes no dataframe vamos usar um metodo chamado .value_counts(), qu conta quantas vezes cada item de uma coluna se repete, que dara como resultado uma series

```python
df['Tipo'].value_counts()
```

Mas como vimos queremos os dados em forma de porcentagem, então colocaremos dentro do parenteses de .value_counts() o metodo **normalize=True** para tornar os dados porcentagem

```python
df['Tipo'].value_counts(normalize=True)
```

Conseguimos as porcentagens, mas para melhorar a visuaização iremos transformar a series em dataframe usando o metodo .to_frame() que será usado sempre que quisermos converter uma series em dataframe

```python
df['Tipo'].value_counts(normalize=True).to_frame()
```

Agora para melhorar iremos organizar de menor para maior percentual usando novamente o .sort_values() colocando o nome da coluna a ser organizada, que nesse caso se tornou 'proportion' pois na nova versão do pandas sempre que uma series for convertida em dataframe a coluna em destaque da series passa a se chamar 'proportion' no dataframe

```python
df['Tipo'].value_counts(normalize=True).to_frame().sort_values('proportion')
```

Para melhorar ainda mais a visuallização faremos um garfico de barras com os dados, salvando o dataframe criado a cima em uma variavel 'df_percentual_tipo'

```python
df_percentual_tipo = df['Tipo'].value_counts(normalize=True).to_frame().sort_values('proportion')
```

Para ajustar essa questão do 'proportions', usaremos o metodo .rename() do pandas para renomear colunas de dataframes e series, que funciona com um dicionario, relacionando o primeiro nome da coluna(o atual) com o que queremos aplicar, alem de usar o metodo **inplace=True** para garantir que a mudança seja aplicada permanente apos a execução do codigo

```python
df_percentual_tipo.rename(columns={'proportion' : 'Percentual'}, inplace=True)
df_percentual_tipo.head()
```

Em seguida criaremos o grafico com o codigo abaixo

*   kind='bar': define o tipo do grafico como grafico de barras
*   figsize=(14,10): define o tamanho do grafico
*   color='green': um pouco auto explicativo, apenas define a cor do grafico
*   edgecolor='black': define a cor das bordas das barras do grafico
*   xlabel='Tipos': define o nome do eixo x
*   ylabel='Percentual': define o nome do eixo y





```python
df_percentual_tipo.plot(kind='bar', figsize=(14, 10), color ='green', edgecolor='black', xlabel = 'Tipos', ylabel = 'Percentual');
```

### **Selecionando apenas os imóveis do tipo apartamento**

Observando o grafico vemos como os apartamentos estão em maioria esmagadora em relação aos outros imoveis, e como vimos que os preços entre casas e apartamentos não são tão distintos na ultima analise, usaremos apenas o tipo 'apartamento' a partir de agora

Pensando nisso vamos utilizar o metodo .query() para selecionar apenas os apartamentos presentes em **df**, analisando se na linha vistada vez por vez pelo metodo(quase como um metodo **for** do python comum) é igual a 'apartamento' na coluna escolhida(Tipo)

```python
df.query('Tipo == "Apartamento"')
```

Agora salvaremos esse novo dataframe que sera utilizado a partir de agora na variavel **df**, alem de usar o .head() apenas para vies de analise se tudo esta correto ou não

# Tratando e filtrando os dados

## Lidando com dados nulos

Geralmente em tratamentos de dados e no trabalho habitual de um cientista de dados, valores nulos em tabelas devem ser tratados para melhor analise e evitar erros ao usar a tabela para algo

De inicio se usa o .isnull() para que seja devolvida uma tabela booleana, ou seja, só tem True(quando o item não é nulo) ou False(quando o item é nulo)

```python
df.isnull()
```

Mas para uma visualização mais ampla, vamos usar o .sum() para recebermos uma series com os valores nulos de cada coluna somados

```python
df.isnull().sum()
```

Agora com uma noção melhor dos valores nulos, vamos substituilos com o metodo .fillna(), que substitui valores nulos pelo numero que esta entre os parenteses, que aqui sera 0(importante lembrar que essa é a melhor abordagem para este projeto, cada projeto tem sua preferência e requisitos)

```python
df.fillna(0)
```

Para garantir que isso seja aplicado de fato, vamos salvar na mesma variavel

```python
df = df.fillna(0)
```

E para confirmar que todos os valores nulos foram retirados, vamos repetir o proceso de usar o .isnull() com o .sum() para recebermos umas series com todos os valores iguas a 0

```python
df.isnull().sum()
```

Agora o que poderia ser um problema futuro para analises mais profundas ou ate em modelos de machine learning que utilizariam esse dataframe, foi agora resolvido

Para viés de curosidade existem outros metodos de lidar com dados nulos, como o **.dropna()**, que remove as linhas que contenham pelo menos um dado nulo, o **.fillna()** ja usado aqui, que substitui o valor nulo pelo numero entre parenteses, porem algo dele que não usamos foi o **method=”ffill”** e o **method=”bfill”**, que respectivamente substituem o valor nulo pelo numero seguinte e pelo numero anterior. Por fim é possivel interpolar dados nulos com calculos baseado nos valores proximos, com o **.interpolate()**

Mas lembre-se de utilisar esses metodos corretamente de acordo com o projeto e seus requisitos

## Removendo registros

Agora teremos a missão de tirar todos os valores inconsistentes da coluna Valor e Condominio, ou seja, todos os valores iguais a zero

Iniciaremos achando tais linhas que queremos tirar com o ja usado metodo query, especificando um "ou" na nossa condição, que é representado na sintaxe como um " | "

```python
df.query("Valor == 0 | Condominio == 0")
```

Agora para fazermos a retirada temos que achar os indices das linhas que desejamos tirar com o .index a frente do codigo anterior

```python
df.query("Valor == 0 | Condominio == 0").index
```

Para facilitarmos o progresso da filtragem dos dados, vamos salvar os indices em uma variavel

```python
registros_a_remover = df.query("Valor == 0 | Condominio == 0").index
```

Para que a remoção seja feita iremos utilisar o metodo .drop(), que dentro de seus parenteses teremos que colocar o que desejamos retirar, no caso a variavel que criamos, porem a uma questão, que é que o metodo .drop() pode retirar tanto linhas quanto colunas, e por isso existe o atributo **axis** que vai dentro do parenteses tambem, onde colocaremos igual a 0 caso queiramos retirar linhas e igual a 1 caso sejam colunas as retiradas, alem disso tem o ja visto inplace=True, que serve para salvar a mudança feita sem a necessidade de salvar em outra variavel

```python
df.drop(registros_a_remover, axis=0, inplace=True)
```

A remoção teoricamente foi feita, vamos confirmar repetindo o .query feito no inicio

```python
df.query("Valor == 0 | Condominio == 0")
```

Foi devolvida uma lista vazia, mostrando que tudo deu certo. Mas sera que nada mais pode ser removido? Veremos agora uma analise do dataframe

```python
df.head()
```

Analisando e lembrando do que ja fizemos anteriormente, vemos que só existe o imovel "Apartamento" na coluna "Tipo", ou seja, a existencia da coluna "Tipo" é irrelevante, vamos remove-la. Para confirmar o que foi dito, vamos usar o metodo .unique() para ver se só "Apartamento" esta na coluna

```python
df['Tipo'].unique()
```

Então para a retirada vamos utilizar novamente o .drop(), passar a coluna "Tipo", colocar agora o metodo **axis** com 1 para retirarmos a coluna apenas e novamente o inplace=True para rapidamente salvar-mos as alterações

```python
df.drop('Tipo', axis=1, inplace=True)
```

Para quitar as duvidas, isso foi feito para simplificar o dataframe retirando informações que não seriam de utilidade, ja que sabemos que apenas apartamentos estão sendo listados, agora para confirmar colocaremos o .head()

```python
df.head()
```

## Filtros

### **1. Apartamentos que possuem `1 quarto` e `aluguel menor que 1200`**



Aqui nos teremos a tarefa de filtrar os dados, que é algo muito simples utilisando pandas, vamos começar encontrando os apartamentos de 1 quarto so

```python
df['Quartos'] == 1
```

Recebemos uma series booleana que diz onde tem ou não o numero 1 na coluna quartos a partir do True ou False(esse proceso poderia muito bem ter sido realizado com .query() tambem, mas foi escolhido este metodo manual), mas queremos uma melhor visualisação, então vamos salvar em uma variavel e então passar para a variavel inicial

```python
selecao1 = df['Quartos'] == 1
df[selecao1]
```

Foi realizada a primeira filtragem, agora nos basta a segunda parte da filtragem que deve estar unida a primeira, mas agora vamos mais direto ao ponto

```python
selecao2 = df['Valor'] < 1200
df[selecao2]
```

Feitas as duas partes da filtragem vamos uni-las agora, para isso vamos criar uma variavel e unir as outras duas com o " & " que significa "and" ou "e" em portugues entre elas

```python
selecao_final = (selecao1) & (selecao2)
df[selecao_final]
```

vamos salvar este ultimo dataframe para que projetos futuros tenham mais facilidade em acessa-lo, ou até nós mesmos

```python
df_1 = df[selecao_final]
```

### **2. `Apartamentos` que possuem pelo menos `2 quartos`, `aluguel menor que 3000` e `area maior que 70`**

Agora temos mais uma filtragem para realisar, um pouco maior mas com os mesmos principios de anteriormente, vamos fazer tudo de uma maneira mais enxuta utilizando todos os conceitos que aprendemos ate agora

```python
selecao = (df['Quartos'] >= 2) & (df['Valor'] < 3000) & (df['Area'] > 70)
df[selecao]
```

Vaos tambem salvar esse codigo pelos mesmos motivos que foram explicados para filtragem anterior

```python
df_2 = df[selecao]
```

# Manipulando os dados

## Criando colunas numéricas

Para concluirmos nossas analises temos mais umas demandas, uma delas é criar duas colunas a mais no dataframe inicial, ou seja, aquele que ainda não tinha sido analisado e ainda tinha por exemplo a coluna tipo. As colunas a serem criadas serão **valor_por_mes** e **valor_por_ano**, referentes aos gastos mensais e anuais para cada imovel

Inicialmente vamos re-importar nosso dataframe inicial

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dado = pd.read_csv(url, sep=";")
dado.head()
```

Em seguida vamos criar **valor_por_mes** com uma simples logica: somar a coluna "Valor" pela coluna "Condomínio". Para criar uma coluna basta colocar como se fossemos chamar uma coluna ja existente, e depois de colocar " = " colocar a condição que definira a coluna

Faremos tambem o bonus de tratar os valores nulos com .fillna() e **inplace=True** como ja foi usado antes

```python
dado.fillna(0, inplace=True)
dado["Valor_por_mes"] = dado['Valor'] + dado['Condominio']
dado.head()
```

Agora criaremos a coluna **valor_por_ano** com a logica de multiplicar a coluna recem criada "Valor_por_mes" por 12 e então somar com a coluna "IPTU"

```python
dado["Valor_por_ano"] = dado["Valor_por_mes"] * 12 + dado["IPTU"]
dado.head()
```

## Criando colunas categóricas

Agora teremos que criar colunas categoricas, no caso duas: "Descrição" e "Possui_suite", sendo que a primeira sera feita a partir de concatenação de strings enquanto o segundo sera booleana, uma coluna binaria

Começando com "Descrição", que é necessário constar o tipo de imóvel, o bairro, a quantidade de quartos e de vagas de garagem, sendo muito semelhante a criação de colunas numericas, porem funcionará com o principio de concatenação

```python
dado["Descrição"] = dado["Tipo"] + dado["Bairro"]
dado[["Descrição"]].head()
```

Note que as informações ficaram coladas umas nas outras, para melhorar isso colocaremos um texto "em" entre a concatenação

```python
dado["Descrição"] = dado["Tipo"] + " em " + dado["Bairro"]
dado[["Descrição"]].head()
```

Agora ficou bem melhor, vamos então colocar as informações que restam usando o conceito de um txto entre a concatenação para completarmos a coluna "Descrição". Lembrar de transformar as colunas contendo inteiros em strings tambem com o metodo .astype() colocando entre parenteses o tipo desejado(no caso "str")

```python
dado['Descrição'] = dado['Tipo'] + ' em ' + dado['Bairro'] + ' com ' + \
                                        dado['Quartos'].astype(str) + ' quarto(s) ' + \
                                        ' e ' + dado['Vagas'].astype(str) + ' vaga(s) de garagem.'
dado.head()
```

Agora criaremos a coluna "Possui_suite" com base na coluna ja existente "Suites" usando o metodo .apply() colocando entre os parenteses uma função que diz que caso o numero de suites seja maior que um, colocaremos "Sim" na nova coluna e "Não" caso a condição não seja atendida, usando uma função lambda para isso

```python
dado['Possui_suite'] = dado['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")
dado.head()
```

Para finalizar vamos salvar para facilitar o acesso futuro para outros projetos

```python
dado.to_csv('dados_completos_dev.csv', index=False, sep=';')
```
